```python
adding_loc_df["Latitude"] = "0"
adding_loc_df["Longitude"] = "0"
```

```python
adding_loc_df.head()
```

```python
# function to select stem fields as a binary
def LATITUDE (row):
   if row['Antenna'] == "SedimentTrapB":
      return '27.326134'
   if row['Antenna'] == "Tanglewood":
      return '27.311757'
   if row['Antenna'] == "Riverview":
      return '27.28313'
   if row['Antenna'] == "Sperandeo":
      return '27.311775'
   if row['Antenna'] == "Byron":
      return '27.309247'
   if row['Antenna'] == "Oskamp":
      return '27.304939'
   if row['Antenna'] == "Duncan":
      return '27.302968'
   if row['Antenna'] == "Tucci1":
      return '27.276147'
   if row['Antenna'] == "Tucci2":
      return '27.276249'
   if row['Antenna'] == "Brown":
      return '27.27506'
   if row['Antenna'] == "Chapman":
      return '27.272518'
   return '0'

def LONGITUDE (row):
   if row['Antenna'] == "SedimentTrapB":
      return '-82.499477'
   if row['Antenna'] == "Tanglewood":
      return '-82.508631'
   if row['Antenna'] == "Riverview":
      return '-82.521402'
   if row['Antenna'] == "Sperandeo":
      return '-82.510186'
   if row['Antenna'] == "Byron":
      return '-82.510443'
   if row['Antenna'] == "Oskamp":
      return '-82.512281'
   if row['Antenna'] == "Duncan":
      return '-82.517611'
   if row['Antenna'] == "Tucci1":
      return '-82.529169'
   if row['Antenna'] == "Tucci2":
      return '-82.526347'
   if row['Antenna'] == "Brown":
      return '-82.526211'
   if row['Antenna'] == "Chapman":
      return '-82.529035'
   return '0'
```

```python
#running the function on the original dataset
adding_loc_df['Latitude'] = adding_loc_df.apply(lambda row: LATITUDE(row), axis=1)
#running the function on the original dataset
adding_loc_df['Longitude'] = adding_loc_df.apply(lambda row: LONGITUDE(row), axis=1)
```

```python
adding_loc_df = adding_loc_df.drop('Antenna', 1)
```

```python
updated_release_df = Release_data_df.join(adding_loc_df)
```

```python
updated_release_df.head()
```

```python
updated_release_df['Latitude'] = pd.to_numeric(updated_release_df['Latitude'],errors='coerce')
updated_release_df['Longitude'] = pd.to_numeric(updated_release_df['Longitude'],errors='coerce')
```

```python
# LOCATION HEATMAP
# Assign the locations to an array of lat and long paires
locations = updated_release_df[["Latitude", "Longitude"]]
#Assign the weights variable to some values
time = updated_release_df["Time.at.Large"]
# Assign the figure variable to the smaps.figures() attribute
fig = gmaps.figure()
# Assign the heatmap_layer variable to the heatmap_layer attributes
heatmap_layer = gmaps.heatmap_layer(locations, weights=time)
# Add the heatmap layer
fig.add_layer(heatmap_layer)
# Call the figure to plot the data
fig
```
